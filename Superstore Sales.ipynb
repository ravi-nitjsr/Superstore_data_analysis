# Categorical Analysis

```python
sales_df.head()
```

## Category wise sales 

```python
a=sns.barplot(x='Category' , y='Sales' , data=sales_df, ci=False )
for i in a.containers:
    a.bar_label(i)
```

```python
sales_df.groupby('Category')['Sales'].mean().plot(kind='pie',autopct='%0.1f%%')
```

```python

sns.set(rc={'figure.figsize':(16,13)})
p=1
for i in ['Ship Mode','Segment','Region','Category',]:
    plt.subplot(3,2,p)
    if i=='Sub-Category':
        ax=sns.countplot(x=i,data=sales_df)
        for i in ax.containers:
            ax.bar_label(i)
        plt.xticks(rotation=90)
    else:
        ax=sns.countplot(x=i,data=sales_df)
        for i in ax.containers:
            ax.bar_label(i)
    p+=1            
```

## sub-category wise sales 

```python
plt.figure(figsize=(12,8))
a=sns.barplot(x='Sub-Category' , y ='Sales' , data = sales_df , ci = False,estimator=sum)
for i in a.containers:
    a.bar_label(i)
plt.xticks(rotation=90);
```

```python
sales_df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
```

```python
sales_df.groupby(['Category','Sub-Category'])['Profit'].mean()
```

```python
sales_df.groupby('Category')['Sub-Category'].value_counts()
```

```python
pd.crosstab(index=sales_df['Category'],columns=sales_df['Sub-Category'], margins =True)
```

```python
dt=sales_df.pivot_table(index='Sub-Category' ,  columns ='Region' ,values="Sales")
dt.style.background_gradient(cmap='viridis')
```

```python
pd.crosstab(index=sales_df['Segment'],columns=sales_df['Ship Mode'],margins=True)
```

```python
sales_df.groupby('Sub-Category')[['Sales','Profit']].mean().head()
```

## Numerical analysis

```python
sns.lmplot(x='Sales',y='Profit',data=sales_df, height=8,aspect=1.25)
```

```python
loss_df=sales_df[sales_df['Profit'] < 0]
loss_df
```

```python
loss_df.groupby(by='Segment').sum()
```

```python
total_loss=np.negative(loss_df['Profit'].sum())
print("Total loss = %.2f" %total_loss)
```

## Time series analysis

```python
sales_df['month']=pd.to_datetime(sales_df['Order Date']).dt.month
sales_df
```

```python
sales_df['year']=pd.to_datetime(sales_df['Order Date']).dt.year
sales_df
```

```python
sales_df['days']=sales_df['Order Date'].apply(lambda x:int(str(x).split('-')[1]))
sales_df
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(x='month',y='Sales',data=sales_df,estimator=sum,ci=False,hue='year',palette='flare')
```
